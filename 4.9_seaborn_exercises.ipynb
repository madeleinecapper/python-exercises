```python
from pydataset import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = data('iris')
iris
```

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
iris
```

```python
sns.distplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?
sns.relplot(data = iris, x='petal_length', y = 'petal_width')
cor_val = iris[['petal_length', 'petal_width']].corr().min()[0]
print(f'String postivie correlation between petal length and petal width at {cor_val}')
```

```python
# Would it be reasonable to predict species 
# based on sepal width and sepal length?
```

```python
sns.relplot(data=iris, x='petal_width', y='sepal_width', hue='species')
```

```python
# we can see visually here that there is a strong positive 
# correlation on sepal_width and petal_width in the versicolor 
# and virginicavspecies. 
# setosa species have a nearly undefined correlation
```

```python
iris[['petal_width', 'sepal_width', 'species']].groupby('species').mean()
```

```python
iris[['petal_width', 'sepal_width', 'species']].groupby('species').max()
```

```python
iris[['petal_width', 'sepal_width', 'species']].groupby('species').min()
```

```python
# When we compare the maximum, minimum, and average petal
# and sepal widths between these species,
# we can confirm our visualization that suggests we have 
# fairly strong data to predict between these species.
# setosa petals have no record of being over 0.6, which is below
# the minimum of the versicolor petal_width.
# Additionally, there is very little overlap between the maximum
# petal width of a versicolor compared to the minimum petal width 
# of a virginica.
```

```python
iris.corr()
```

```python
sns.pairplot(iris, hue='species')
```

```python
# When we pairplot all of our data from iris and visualize
# it with a hue by species, we can see distinct groupings in 
# many of the plots.  When looking at the data types individually,
# we see that petal_length and petal_width fall into the most
# distinct separate bins.  When we combine these and view the 
# plot of these two features with our hue, we can see disctinctly 
# separate bins.
sns.relplot(data=iris, x='petal_width', y='petal_length', hue='species')
```

```python
iris[['species','petal_length','petal_width']].groupby('species').mean()
```

```python
trimmed_data = iris[['species','petal_length','petal_width']].groupby('species').mean()
sns.heatmap(trimmed_data, annot=True)
```

## Using the chipotle dataset from the previous exercise, create a barplot that shows the 4 most popular items and the revenue produced by each.

```python
def get_engine(db, user, host, password):
    from sqlalchemy import create_engine
    url = f'mysql+pymysql://{user}:{password}@{host}/{db}'
    return create_engine(url)
from env import user, host, pw
conn = get_engine('chipotle', user, host, pw)
chipotle = pd.read_sql('SELECT * FROM orders', conn)
```

```python
chipotle
```

```python
four_most_pop = chipotle[['item_name', 'quantity']].groupby('item_name', as_index=False).sum().sort_values(by='quantity', ascending=False).head(4)
```

```python
four_most_pop
```

```python
sns.barplot(x='item_name', y='quantity', data=four_most_pop)
```

```python
sns.lineplot(x='Subject', y='diff_over_time', data = reac_diff)
```
