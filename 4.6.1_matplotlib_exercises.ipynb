## Use matplotlib to plot y = x^2 - x + 2

```python
import matplotlib.pyplot as plt
x = range(-50,50)
y = [n ** 2 - n + 2 for n in x]
plt.text(0, -100, '(0, 0)', fontsize = 10, color = 'blue')
plt.plot(x, y)
plt.show()
```

## Combine the figures you created in the last step into one large figure with 4 subplots

```python
from math import tan
x1 = range(0,20)
x2 = range(-10,10)
x3 = range(-7, 7)
x4 = range(-10,10)
y1 = [n ** 0.5 for n in x1]
y2 = [n ** 3 for n in x2]
y3 = [tan(n) for n in x3]
y4 = [2 ** n for n in x4]

plt.subplot(221)
plt.plot(x1,y1)
plt.title('sqrt(x)')
plt.subplot(222)
plt.plot(x2,y2)
plt.title('x^3')
plt.subplot(223)
plt.plot(x3,y3)
plt.title('tan(x)')
plt.subplot(224)
plt.plot(x4,y4)
plt.title('2^x')
plt.suptitle('some plots')

plt.show()
```

```python
from math import tan
x1 = range(0,10)
x2 = range(-10,10)
x3 = range(-10, 10)
x4 = range(-10,10)
y1 = [n ** 0.5 for n in x1]
y2 = [n ** 3 for n in x2]
y3 = [tan(n) for n in x3]
y4 = [2 ** n for n in x4]
plt.figure(figsize = (13, 8))
plt.plot(x1,y1, c= 'magenta', label = 'sqrt(x)')
plt.plot(x2,y2, c = 'green', label = 'x^3')
plt.plot(x3,y3, c = 'turquoise', label = 'tan(x)')
plt.plot(x4,y4, c = 'plum', label = '2^x')
plt.legend(loc = 'upper left')
plt.title('I\'m sorry')
plt.show()
```
